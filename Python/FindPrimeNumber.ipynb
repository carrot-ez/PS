```
def permutation(arr, n, isVisit, curr, result) :
    if len(curr) == n :
        tmp = ""
        for e in curr :
            tmp += str(e)
        result.append(int(tmp));
        return;
    
    for i in range(len(arr)) :
        if isVisit[i] == 1 :
            continue
        isVisit[i] = 1
        curr.append(arr[i])

        permutation(arr, n, isVisit, curr, result)

        isVisit[i] = 0
        curr.pop()
```

```
number_set = set()
def permutation(base, array):
    if base:
        number_set.add(int(base))

    for i, s in enumerate(array):
        permutation(base + s, array[:i] + array[i+1:])
```
